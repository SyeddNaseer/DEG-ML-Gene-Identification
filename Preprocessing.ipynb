```python
from bioservices import Ensembl

# Initialize Ensembl service
ensembl = Ensembl()

# List of Ensembl Gene IDs
ensembl_ids = [
    "ENSG00000201659", "ENSG00000059122", "ENSG00000207656",
    "ENSG00000211513", "ENSG00000199109", "ENSG00000207962",
    "ENSG00000283891", "ENSG00000210194", "ENSG00000199047",
    "ENSG00000199088"
]

# Function to fetch gene name
def get_gene_name(ensembl_id):
    try:
        # Fetch gene information using get_gene_name method
        gene_info = ensembl.get_gene_name(ensembl_id)
        gene_name = gene_info.get('display_name', 'N/A')
        return gene_name
    except Exception as e:
        print(f"Error retrieving data for {ensembl_id}: {e}")
        return 'N/A'

# Get gene names
gene_names = {ensembl_id: get_gene_name(ensembl_id) for ensembl_id in ensembl_ids}

# Print the results
for ensembl_id, gene_name in gene_names.items():
    print(f"{ensembl_id}: {gene_name}")
```
