```python
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler

# Load the CSV file
df = pd.read_csv("transposed2.csv")

# Separate the features (genes) and label
features = df.drop(columns=['Geneid', 'Label'])
labels = df['Label']

# Imputation of missing values (mean imputation)
imputer = SimpleImputer(strategy='mean')
features_imputed = imputer.fit_transform(features)

# Standardization
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features_imputed)

# Convert the result back to a DataFrame and combine with Geneid and Label
features_scaled_df = pd.DataFrame(features_scaled, columns=features.columns)
final_df = pd.concat([df[['Geneid']], features_scaled_df, labels], axis=1)

# Save the standardized data to a new CSV file
final_df.to_csv("transposed_scaled.csv", index=False)

# Show the standardized data (optional)
print(final_df)
```
